First decode the large grib into smaller single variable gribs

```python
variables = ['var165', 'var166', 'var167', 'var151', 'var34', 'var134']
```

Extract the variable names as CDO loses them

```python
names={}
for var in variables:
    x = xr.open_dataset('/mnt/kingfast1TB/out/'+str(var)+'.grb')
    names[var] = list(x.data_vars.keys())[0]
```
